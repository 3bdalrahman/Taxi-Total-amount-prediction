### MLP

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping

preprocessor = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown='ignore'), ['passenger_count', 'payment_type', 'pickup_weekday']),
], remainder='passthrough')

X_processed = preprocessor.fit_transform(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_processed)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define the neural network model
model = Sequential([
    Dense(256, activation='relu'),
    BatchNormalization(),

    Dense(128, activation='relu'),
    BatchNormalization(),


    Dense(64, activation='relu'),
    BatchNormalization(),

    Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mse', metrics=[tf.keras.metrics.RootMeanSquaredError()])

# Define EarlyStopping callback
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model 
history = model.fit(X_train, y_train, validation_split=0.1, epochs=50, batch_size=1024, callbacks=[early_stop])

loss, rmse = model.evaluate(X_test, y_test)
print(f"\nTest RMSE: {rmse}")

y_pred = model.predict(X_test).flatten()
r2 = r2_score(y_test, y_pred)
print(f"R² Score: {r2}")
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
rmse = history.history['root_mean_squared_error']
val_rmse = history.history['val_root_mean_squared_error']

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.title('Training vs Validation Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(rmse, label='Training RMSE')
plt.plot(val_rmse, label='Validation RMSE')
plt.xlabel('Epochs')
plt.ylabel('RMSE')
plt.title('Training vs Validation RMSE')
plt.legend()

plt.tight_layout()
plt.show()
```

### TabNet

```python
from pytorch_tabnet.tab_model import TabNetRegressor
import torch

preprocessor = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown='ignore'), ['passenger_count', 'payment_type', 'pickup_weekday']),
], remainder='passthrough')

X_processed = preprocessor.fit_transform(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_processed)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Convert data to proper shape and type
X_train = X_train.astype(np.float32)
X_test = X_test.astype(np.float32)
y_train = y_train.values.reshape(-1, 1).astype(np.float32)
y_test = y_test.values.reshape(-1, 1).astype(np.float32)

# Define TabNet model
tabnet_model = TabNetRegressor(
    optimizer_fn=torch.optim.Adam,
    optimizer_params=dict(lr=2e-2),
    scheduler_params={"step_size": 10, "gamma": 0.9},
    scheduler_fn=torch.optim.lr_scheduler.StepLR,
    mask_type='sparsemax',
    device_name='cpu'
)

# Train the model
tabnet_model.fit(
    X_train=X_train, y_train=y_train,
    eval_set=[(X_test, y_test)],
    eval_metric=['rmse'],
    max_epochs=50,
    patience=5,
    batch_size=1024,
    virtual_batch_size=128,
    num_workers=0,
    drop_last=False
)

y_pred = tabnet_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"\nTest RMSE: {rmse}")
print(f"R² Score: {r2}")
```

```python
import matplotlib.pyplot as plt

loss = tabnet_model.history['loss']
val_rmse = tabnet_model.history['val_0_rmse']

plt.figure(figsize=(12, 6))

# Loss curve
plt.subplot(1, 2, 1)
plt.plot(loss, label='Train Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train Loss per Epoch')
plt.legend()
plt.grid(True)

# RMSE curve
plt.subplot(1, 2, 2)
plt.plot(val_rmse, label='Validation RMSE', color='orange')
plt.xlabel('Epoch')
plt.ylabel('RMSE')
plt.title('Validation RMSE per Epoch')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

### Wide & Deep

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Concatenate, BatchNormalization
from tensorflow.keras.models import Model

cat_features = ['passenger_count', 'payment_type', 'pickup_weekday']

preprocessor = ColumnTransformer([
    ("wide", OneHotEncoder(handle_unknown='ignore'), cat_features),
], remainder='passthrough')

X_processed = preprocessor.fit_transform(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_processed)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# convert to float32
X_train = X_train.astype(np.float32)
X_test = X_test.astype(np.float32)
y_train = y_train.astype(np.float32)
y_test = y_test.astype(np.float32)

# ========== Build Wide & Deep Model ==========
input_layer = Input(shape=(X_train.shape[1],))

# Wide part: Linear layer (no activation)
wide_output = Dense(1)(input_layer)

# Deep part: Feedforward layers
deep = Dense(256, activation='relu')(input_layer)
deep = BatchNormalization()(deep)
deep = Dense(128, activation='relu')(deep)
deep = BatchNormalization()(deep)
deep = Dense(64, activation='relu')(deep)
deep = BatchNormalization()(deep)
deep_output = Dense(1)(deep)

# Combine Wide + Deep outputs
combined = Concatenate()([wide_output, deep_output])
final_output = Dense(1)(combined)

# Define model
model = Model(inputs=input_layer, outputs=final_output)

# Compile
model.compile(optimizer='adam', loss='mse', metrics=[tf.keras.metrics.RootMeanSquaredError()])

# Early stopping
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train
history = model.fit(X_train, y_train, validation_split=0.1, epochs=50, batch_size=1024, callbacks=[early_stop])

loss, rmse = model.evaluate(X_test, y_test)
print(f"\nTest RMSE: {rmse}")

y_pred = model.predict(X_test).flatten()
r2 = r2_score(y_test, y_pred)
print(f"R² Score: {r2}")
```

```python
import matplotlib.pyplot as plt

loss = history.history['loss']
val_loss = history.history['val_loss']
rmse = history.history['root_mean_squared_error']
val_rmse = history.history['val_root_mean_squared_error']

plt.figure(figsize=(12, 5))

# Loss plot
plt.subplot(1, 2, 1)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.title('Training vs Validation Loss')
plt.legend()

# RMSE plot
plt.subplot(1, 2, 2)
plt.plot(rmse, label='Training RMSE')
plt.plot(val_rmse, label='Validation RMSE')
plt.xlabel('Epochs')
plt.ylabel('RMSE')
plt.title('Training vs Validation RMSE')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTETomek

X_sampled = X_with_clusters.sample(n=1000000, random_state=42)
y_sampled = y.loc[X_sampled.index]

bins = [0, 10, 20, 30, 50, 250]
labels = [0, 1, 2, 3, 4]
y_binned = pd.qcut(y_sampled, q=5, labels=labels)

X_train, X_test, y_train_binned, y_test_binned = train_test_split(X_sampled, y_binned, test_size=0.2, random_state=42)

categorical_cols = ['passenger_count', 'payment_type', 'pickup_weekday']
preprocessor = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown='ignore'), categorical_cols)
], remainder='passthrough')

model_pipeline = ImbPipeline([
    ("pre", preprocessor),
    ("smote", SMOTETomek(random_state=42)),
    ("clf", LogisticRegression(max_iter=200, solver='liblinear', class_weight='balanced'))
])

model_pipeline.fit(X_train, y_train_binned)

y_pred_binned = model_pipeline.predict(X_test)

cm = confusion_matrix(y_test_binned, y_pred_binned)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot(cmap='Blues')
plt.title("Confusion Matrix (Binned Total Amount)")
plt.show()

print("Classification Report:")
print(classification_report(y_test_binned, y_pred_binned))
```
